### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/TheGabrielVieira/challenge1-data-science-main/master/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/TheGabrielVieira/challenge1-data-science-main/master/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/TheGabrielVieira/challenge1-data-science-main/master/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/TheGabrielVieira/challenge1-data-science-main/master/base-de-dados-challenge-1/loja_4.csv"


loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

# 1. Análise do faturamento


```python
# Calculando o faturamento total de cada loja
faturamento_1 = loja1["Preço"].sum()
faturamento_2 = loja2["Preço"].sum()
faturamento_3 = loja3["Preço"].sum()
faturamento_4 = loja4["Preço"].sum()

# Exibindo os resultados
print("Faturamento Loja 1: R$", round(faturamento_1, 2))
print("Faturamento Loja 2: R$", round(faturamento_2, 2))
print("Faturamento Loja 3: R$", round(faturamento_3, 2))
print("Faturamento Loja 4: R$", round(faturamento_4, 2))

# Faturamento total geral (opcional)
faturamento_total = faturamento_1 + faturamento_2 + faturamento_3 + faturamento_4
print("\nFaturamento Total (Todas as Lojas): R$", round(faturamento_total, 2))
```

```python
import matplotlib.pyplot as plt

# Dados do faturamento
faturamentos = [faturamento_1, faturamento_2, faturamento_3, faturamento_4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Criando o gráfico
plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamentos, color='royalblue')
plt.title('Faturamento por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Exibindo o gráfico
plt.show()
```

# 2. Vendas por Categoria


```python
categorias_loja1 = loja1.groupby('Categoria do Produto').size().sort_values(ascending=False)
categorias_loja2 = loja2.groupby('Categoria do Produto').size().sort_values(ascending=False)
categorias_loja3 = loja3.groupby('Categoria do Produto').size().sort_values(ascending=False)
categorias_loja4 = loja4.groupby('Categoria do Produto').size().sort_values(ascending=False)

print("Loja 1 - Vendas por Categoria:")
print(categorias_loja1, end="\n\n")

print("Loja 2 - Vendas por Categoria:")
print(categorias_loja2, end="\n\n")

print("Loja 3 - Vendas por Categoria:")
print(categorias_loja3, end="\n\n")

print("Loja 4 - Vendas por Categoria:")
print(categorias_loja4)
```

```python
import matplotlib.pyplot as plt

# Função para plotar gráfico de barras
def plot_vendas_por_categoria(categorias, titulo):
    plt.figure(figsize=(10, 5))
    categorias.plot(kind='bar')
    plt.title(titulo)
    plt.xlabel('Categoria do Produto')
    plt.ylabel('Número de Vendas')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Plotando os gráficos
plot_vendas_por_categoria(categorias_loja1, 'Loja 1 - Vendas por Categoria')
plot_vendas_por_categoria(categorias_loja2, 'Loja 2 - Vendas por Categoria')
plot_vendas_por_categoria(categorias_loja3, 'Loja 3 - Vendas por Categoria')
plot_vendas_por_categoria(categorias_loja4, 'Loja 4 - Vendas por Categoria')
```

```python
# Unindo os dados em um único DataFrame para comparação
comparacao_df = pd.DataFrame({
    'Loja 1': categorias_loja1,
    'Loja 2': categorias_loja2,
    'Loja 3': categorias_loja3,
    'Loja 4': categorias_loja4
}).fillna(0).astype(int)  # Preenche categorias ausentes com 0 e converte para int

# Plotando gráfico de barras agrupadas
plt.figure(figsize=(12, 6))
comparacao_df.plot(kind='bar', width=0.8)
plt.title('Comparação de Vendas por Categoria entre Lojas')
plt.xlabel('Categoria do Produto')
plt.ylabel('Número de Vendas')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python

avaliacao_loja1 = loja1.groupby('Avaliação da compra').size().sort_values(ascending=False)
avaliacao_loja2 = loja2.groupby('Avaliação da compra').size().sort_values(ascending=False)
avaliacao_loja3 = loja3.groupby('Avaliação da compra').size().sort_values(ascending=False)
avaliacao_loja4 = loja4.groupby('Avaliação da compra').size().sort_values(ascending=False)

print("Loja 1 - Avaliação da compra:")
print(avaliacao_loja1, end="\n\n")

print("Loja 2 - Avaliação da compra:")
print(avaliacao_loja2, end="\n\n")

print("Loja 3 - Avaliação da compra:")
print(avaliacao_loja3, end="\n\n")

print("Loja 4 - Avaliação da compra:")
print(avaliacao_loja4)

lojas = {
    'Loja 1': loja1,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

print("Média de avaliação por loja:")
for nome_loja, df_loja in lojas.items():
    media_avaliacao = df_loja["Avaliação da compra"].mean()
    print(f"- {nome_loja}: {media_avaliacao:.2f}")
```

```python
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [
    loja1["Avaliação da compra"].mean(),
    loja2["Avaliação da compra"].mean(),
    loja3["Avaliação da compra"].mean(),
    loja4["Avaliação da compra"].mean()
]

# Criando o gráfico de barras
plt.figure(figsize=(8, 6)) # Define o tamanho da figura
plt.bar(nomes_lojas, medias, color=['lightcoral', 'lightgreen', 'lightskyblue', 'lightsalmon'])

# Adicionando título e rótulos
plt.title('Média de Avaliação da Compra por Loja', fontsize=16)
plt.xlabel('Loja', fontsize=12)
plt.ylabel('Média de Avaliação', fontsize=12)
plt.ylim(3.8, 4.1) # Define os limites do eixo y para melhor visualização das diferenças
plt.grid(axis='y', linestyle='--')

# Adicionando os valores das médias acima das barras
for i, v in enumerate(medias):
    plt.text(i, v + 0.01, f'{v:.2f}', ha='center', va='bottom')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Adicionar coluna identificando a loja
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatenar os dados
dados = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# Contar quantas vezes cada produto foi vendido por loja
vendas_por_produto_loja = dados.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade Vendida')

# Identificar o produto mais e o menos vendido por loja
mais_vendidos = vendas_por_produto_loja.loc[vendas_por_produto_loja.groupby('Loja')['Quantidade Vendida'].idxmax()]
menos_vendidos = vendas_por_produto_loja.loc[vendas_por_produto_loja.groupby('Loja')['Quantidade Vendida'].idxmin()]

# Unir resultados
destaques_vendas = pd.concat([
    mais_vendidos.assign(Destaque='Mais Vendido'),
    menos_vendidos.assign(Destaque='Menos Vendido')
])

destaques_vendas.sort_values(['Loja', 'Destaque'], inplace=True)
destaques_vendas.reset_index(drop=True, inplace=True)
destaques_vendas
```

```python
import seaborn as sns

# Configurar estilo do gráfico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=destaques_vendas,
    x='Loja',
    y='Quantidade Vendida',
    hue='Produto',
    palette='viridis',
    dodge=True
)

plt.title('Produtos Mais e Menos Vendidos por Loja')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Loja')
plt.legend(title='Produto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 5. Frete Médio por Loja


```python
# Calcular o frete médio por loja
frete_medio_loja1 = loja1['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Exibir os resultados
print(f"Custo médio de frete - Loja 1: R${frete_medio_loja1:.2f}")
print(f"Custo médio de frete - Loja 2: R${frete_medio_loja2:.2f}")
print(f"Custo médio de frete - Loja 3: R${frete_medio_loja3:.2f}")
print(f"Custo médio de frete - Loja 4: R${frete_medio_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cálculo dos fretes médios
fretes_medios = {
    'Loja 1': loja1['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean(),
}

# Converter para DataFrame para facilitar visualização
fretes_df = pd.DataFrame(list(fretes_medios.items()), columns=['Loja', 'Frete Médio'])

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=fretes_df, x='Loja', y='Frete Médio', hue='Loja', palette='Blues_d', legend=False)

plt.title('Custo Médio de Frete por Loja')
plt.ylabel('Frete Médio (R$)')
plt.xlabel('Loja')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Relatório de Análise: Recomendação de Venda de Loja para o Senhor João
Introdução
O objetivo deste relatório é apoiar o Senhor João na decisão de qual de suas quatro lojas deve ser vendida. Para isso, realizamos uma análise detalhada dos principais indicadores de desempenho de cada loja, incluindo o faturamento total, desempenho por categoria de produtos, avaliações dos clientes, produtos mais e menos vendidos, além do custo médio de frete. Com base nessas métricas, propomos uma recomendação clara e justificada da loja menos estratégica para manutenção no portfólio do Senhor João.

Desenvolvimento
1. Faturamento Total por Loja
O faturamento representa o desempenho bruto de vendas e, portanto, é um dos principais indicadores da saúde financeira de uma loja. Veja o resumo abaixo:

Loja 1: R$ 1.534.509,12

Loja 2: R$ 1.488.459,06

Loja 3: R$ 1.464.025,03

Loja 4: R$ 1.384.497,58

Análise: A Loja 1 lidera em faturamento, seguida pela Loja 2. A Loja 4 apresenta o menor faturamento entre as quatro.

2. Vendas por Categoria de Produtos
As lojas compartilham categorias semelhantes, com destaque para móveis e eletrônicos, que são consistentemente os mais vendidos.

Mais vendida (todas as lojas): Móveis

Menos vendida por loja:

Loja 1: Utilidades domésticas

Loja 2: Utilidades domésticas

Loja 3: Instrumentos musicais

Loja 4: Instrumentos musicais

Análise: A Loja 1 e a Loja 2 possuem boa distribuição nas categorias de maior demanda. Já a Loja 4 possui uma performance menos equilibrada, com algumas categorias com baixo giro.

3. Avaliação Média dos Clientes
A satisfação do cliente é fundamental para a sustentabilidade do negócio. Veja as médias de avaliação (de 1 a 5):

Loja 3: 4.05

Loja 2: 4.04

Loja 4: 4.00

Loja 1: 3.98

Análise: A Loja 3 tem o melhor índice de satisfação, o que pode indicar fidelidade e melhor experiência do cliente. A Loja 1, apesar de líder em faturamento, apresenta a pior avaliação.

4. Produtos Mais e Menos Vendidos
A análise dos produtos mais e menos vendidos ajuda a identificar o foco e os gargalos operacionais.

Todas as lojas apresentam produtos populares distintos, mas também revelam produtos com baixa saída, o que pode indicar problemas de mix de estoque ou segmentação inadequada.

Análise: Não há uma loja com extremo problema neste quesito, mas lojas com maior diversidade (como a Loja 3) demonstram maior resiliência de vendas.

5. Custo Médio de Frete
Um frete mais barato pode indicar maior lucratividade líquida por venda.

Loja 4: R$ 31,28

Loja 3: R$ 33,07

Loja 2: R$ 33,62

Loja 1: R$ 34,69

Análise: A Loja 4 tem o menor custo médio de frete, o que poderia ser uma vantagem competitiva. No entanto, seu baixo faturamento e menor destaque nos demais indicadores enfraquecem essa vantagem.

Conclusão e Recomendação
Com base na análise de todos os indicadores avaliados — faturamento, desempenho por categoria, avaliações dos clientes, produtos mais/menos vendidos e frete — recomendamos que o Senhor João venda a Loja 4.

Apesar do frete mais barato, a Loja 4 apresenta o menor faturamento, menor equilíbrio entre as categorias de produtos, e avaliação média inferior às Lojas 2 e 3. Ela também não lidera em nenhuma das métricas principais. Por outro lado, as Lojas 1, 2 e 3 apresentam maior robustez comercial, seja pelo faturamento, reputação com os clientes ou diversidade de vendas.

Portanto, vender a Loja 4 representa a decisão mais estratégica, liberando recursos e esforços que podem ser realocados para as lojas com melhor desempenho e maior potencial de crescimento.
